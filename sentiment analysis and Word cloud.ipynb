### Scraping data from Skytrax

If you visit [https://www.airlinequality.com] you can see that there is a lot of data there. For this task, we are only interested in reviews related to British Airways and the Airline itself.

If you navigate to this link: [https://www.airlinequality.com/airline-reviews/british-airways] you will see this data. Now, we can use `Python` and `BeautifulSoup` to collect all the links to the reviews and then to collect the text data on each of the individual review links.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
plt.style.use('ggplot')
```

```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 10
page_size = 100

reviews = []
score = []

# for i in range(1, pages + 1):
for i in range(1, pages + 1):

    print(f"Scraping page {i}")

    # Create URL to collect links from paginated data
    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"

    # Collect HTML data from this page
    response = requests.get(url)

    # Parse content
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')
    review_rating_divs = parsed_content.find_all('div', itemprop='reviewRating')
    
    for para in parsed_content.find_all("div", {"class": "text_content"}):
        reviews.append(para.get_text())    
    for div in review_rating_divs:
        score.append(div.find('span', itemprop='ratingValue').get_text())

    print(f"   ---> {len(reviews), len(score)} total reviews & score")
```

```python
df = pd.DataFrame()
df["reviews"] = reviews
df["score_out_10"] = score
df.head()
```

Congratulations! Now you have your dataset for this task! The loops above collected 1000 reviews by iterating through the paginated pages on the website. However, if you want to collect more data, try increasing the number of pages!

 The next thing that you should do is clean this data to remove any unnecessary text from each of the rows. For example, "✅ Trip Verified" can be removed from each row if it exists, as it's not relevant to what we want to investigate.

```python
def remove_trip_verified(df):
    df['reviews'] = df['reviews'].str.replace('✅ Trip Verified | ', '')
    df['reviews'] = df['reviews'].str.replace('Not Verified | ', '')
    return df

df = remove_trip_verified(df.copy())

print(df)
```

## Text preprocessing

```python
import re
```

### 1. Converting our documents into lower case

```python
reviews_lc = []

# Convert and append each element to the lowercase_list
for doc in reviews:
    reviews_lc.append(doc.lower())
reviews_lc
```

```python
# Remove the specified substrings from each element "✅ Trip Verified |" & "Not Verified |".

a="✅ trip verified |"
b="not verified |"
#reviews_lc = [doc.replace(a, "").replace(b, "") for doc in reviews_lc]
#reviews_lc

reviews_lc1=[]
for doc in reviews_lc:
    reviews_lc1.append(doc.replace(a, "").replace(b, ""))
```

```python
## remove punctuation marks

import string

# Step 1: Create a translation table to remove punctuation and special characters
translator = str.maketrans('', '', string.punctuation)
# Step 2: Remove punctuation and special characters from each document
reviews_lc1 = [doc.translate(translator) for doc in reviews_lc1]
```

```python
# remove non-breaking space "\xa0", "\r" & "\n", and commas
reviews_lc1 = [doc.replace("\xa0"," ").replace("\r"," ").replace("\n"," ").replace("’"," ") for doc in reviews_lc1]
```

```python
# removing repeatetive words related to company British Airways
variations = [r'British Airways',r'british airways',r'BritishAirways',r'britishairways',r'British Air',r'british air',
              r'british airway',r'British Airway',r'BritishAirway',r'britishairway', r'ba',r'BA',r'b a', r'B A']
pattern = '|'.join(variations)
cleaned_documents = []
for document in reviews_lc1:
    cleaned_document = re.sub(pattern, '', document)
    cleaned_documents.append(cleaned_document)
```

```python
cleaned_documents
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')# Download the stop words if not already downloaded
```

```python
final_documents = []
for x in cleaned_documents:
    filtered_words=[]
    words= nltk.word_tokenize(x)
    filtered_words = [word for word in words if word.lower() not in stopwords.words('english')]
    filtered_text = ' '.join(filtered_words)
    final_documents.append(filtered_text)
```

```python
len(final_documents)
```

```python
new_df = pd.DataFrame()
```

```python
new_df["cleaned_reviews"]= final_documents
new_df
```

```python
from nltk.sentiment import vader
nltk.download('vader_lexicon')
```

```python
# Initiating an instance of SentimentIntensityAnalyzer
sentiment_analyzer = vader.SentimentIntensityAnalyzer()
```

```python
# fiting the vader analyser on text and creating a new column named 'SENTIMENT'
new_df['SENTIMENT'] = new_df['cleaned_reviews'].apply(lambda review: sentiment_analyzer.polarity_scores(review)['compound'])
new_df['Neg'] = new_df['cleaned_reviews'].apply(lambda review: sentiment_analyzer.polarity_scores(review)['neg'])
new_df['Neu'] = new_df['cleaned_reviews'].apply(lambda review: sentiment_analyzer.polarity_scores(review)['neu'])
new_df['Pos'] = new_df['cleaned_reviews'].apply(lambda review: sentiment_analyzer.polarity_scores(review)['pos'])

new_df
```

```python
# Creating another new column called 'SENTIMENT_TYPE' that contains the type of sentiment of each review using nltk
# for each sentiment score in the 'SENTIMENT' column
new_df["Sentiment_cat"] = new_df["SENTIMENT"].apply(lambda x: "Positive" if x > 0 else ("Negative" if x < 0 else "Neutral"))
new_df
```

```python
# calculate the percentage of each type of sentiment score
sent_cat_percentage = new_df['Sentiment_cat'].value_counts(normalize=True) * 100

# create the pie chart
fig, ax = plt.subplots()
ax.pie(
    sent_cat_percentage,
    labels=sent_cat_percentage.index, 
    colors=['green', 'red', 'grey'], 
    explode=[0.1, 0, 0], 
    autopct='%1.1f%%'
    )

plt.show()
```

## Word cloud (Extracting the top 20 words which are being talked about)

```python
#nltk.download('stopwords')
from collections import Counter
```

```python
# The frequency chart shows the key topics within the 
# reviews showing what people are actively talking about their experience and staff

# Get the reviews from the dataframe
reviews = new_df['cleaned_reviews']

# Convert the reviews to a list of words
words = []
for review in reviews:
    words.extend(review.split())

# Filter out the stubborn stop words that don't want to disappear
stopwords = nltk.corpus.stopwords.words('english')
key_words = [word for word in words if word not in stopwords and word not in ['I', 'The','would','one','get','-']]


# Get the top 20 most used key words
word_counts = Counter(key_words)
top_20_words = word_counts.most_common(20)

# Filter out the stop words 
filtered_words = [
    (word, count) for word, count in top_20_words if word not in stopwords] #and word not in ['I', 'The','would','one','get','-']]

# Get the labels and values for the histogram
labels, values = zip(*top_20_words)

# Create the histogram
plt.bar(labels, values)

# Set the x-axis labels to be vertical
plt.xticks(rotation=90)

#Display the plot
plt.show()
```
